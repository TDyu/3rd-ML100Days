```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
(row_count, column_count) = app_train.shape
row_count, column_count
```

#### 列出所有欄位

```python
for col in app_train.columns: 
    print(col) 
```

#### 截取部分資料

```python
# 以 column name 取出單行成 Series
app_train.loc[:, 'NAME_CONTRACT_TYPE']
```

```python
# 以 column name 取出多行成 DF
app_train.loc[:, 'NAME_CONTRACT_TYPE':'CODE_GENDER']
```

```python
# 以 index number 取出單行成 Series
app_train.iloc[:, 2]
```

```python
# 以 index number 取出多行成 DF
app_train.iloc[:, [2,3]] # = app_train.iloc[:, 2:3]
```
